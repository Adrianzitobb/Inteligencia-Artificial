```python
# Paso 1: Cargar librerías
import matplotlib.pyplot as plt # Generar gráficas y dibujos
import numpy as np # Generar arreglos con operaciones matemáticas
import pandas as pd # Cargar conjuntos de datos externos
import sklearn.linear_model # Contiene diferentes modelos de ML
```

```python
# Paso 2: Cargar datos

np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 10 * horas_estudio + np.random.randn(100, 1)
```

```python
# Paso 3: Preprocesamiento de datos


from sklearn.model_selection import train_test_split #Dividir datos de entrenamiento y datos de prueba
train_data, train_label, test_data, test_label = train_test_split(horas_estudio, calificaciones, test_size=0.2, random_state=0)
```

```python
# Paso 4: Selección y entrenamiento del modelo

model = sklearn.linear_model.LinearRegression()
model.fit(horas_estudio, calificaciones)
```

```python
# Paso 5: Prueba del modelo

horas_estudio_nuevo = [[2.9685]] # Predicciones con el modelo entrenado (1)
print(model.predict(horas_estudio_nuevo))

calificaciones_nuevo = [[3.0]] # Predicciones con el modelo entrenado (2)
print(model.predict(calificaciones_nuevo))

plt.scatter(horas_estudio, calificaciones, color='red') # Gráfica final con el modelo entrenado
plt.xlabel("HORAS DE ESTUDIO")
plt.ylabel("CALIFICACIONES")
plt.title("EMC con regresión lineal")
plt.ylim(0, 100)
plt.plot(horas_estudio, model.predict(horas_estudio), color='blue')
```
